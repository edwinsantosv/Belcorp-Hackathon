# Catalogues

```python
url_catalogues="https://api-qa.belcorp.biz/catalogs/PE"
response_catalogues = requests.request("GET", url_catalogues, headers=headers, data=payload)

print(response_catalogues.text)
```

```python
data_catalogues = json.loads(response_catalogues.text)["results"]
df_catalogues = pd.DataFrame(data_catalogues)
```

```python
# Mostrar DataFrame
df_catalogues.head()
```

```python
from PIL import Image
from io import BytesIO
```

```python
url=df_catalogues.cover_image[1]
```

```python
# Descargar la imagen desde la URL
picture = requests.get(url)
```

```python
if picture.status_code == 200:
    # Convertir el contenido a una imagen usando Pillow
    img = Image.open(BytesIO(picture.content))
    #img.save("downloaded_image.jpg", "JPEG")
    # Mostrar la imagen en el notebook
    display(img)
else:
    print(f"Error al descargar la imagen: {picture.status_code}")
```
