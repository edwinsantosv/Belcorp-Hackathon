## All Catalogues

```python
def get_catalogues(country,headers=headers,payload=payload):
    url_catalogues=f"https://api-qa.belcorp.biz/catalogs/{country}"
    response_catalogues = requests.request("GET", url_catalogues, headers=headers, data=payload)
    data_catalogues = json.loads(response_catalogues.text)["results"]
    df_catalogues = pd.DataFrame(data_catalogues)

    return df_catalogues
```
